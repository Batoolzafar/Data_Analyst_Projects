```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv("housing.csv")
```

```python
file.columns
```

### Change in price with time

```python
sns.relplot(data=file, x="Yr Sold", y="SalePrice", kind="line")
```

### Relation between Sale price and Ground Living Area

```python
sns.scatterplot(file,x= "Gr Liv Area",y = "SalePrice")
plt.show()   # relation follows a trend except the three outliers
```

```python
sns.relplot(file,x= "Gr Liv Area",y = "SalePrice",col = "Neighborhood" ) 
plt.show()
# the neighborhoods with houses of similar area and price may have people with same lifestyle and comparable income
# the outliers belong to the same neighborhood "Edwards" 
# the houses in that neighborhood are low in prices, the three outliers may be following the price trend in neighborhood
```

```python
sns.scatterplot(file,x= "Gr Liv Area",y = "SalePrice", hue = "Overall Qual" ) 
# Increasing price increases quality
```

```python
sns.scatterplot(file,x= "Gr Liv Area",y = "SalePrice", hue = "Garage Cars" )
# large houses more garage cars
```

```python
sns.relplot(file,x= "Gr Liv Area",y = "SalePrice",col = "Neighborhood",hue = "Garage Cars" ,style =  "Garage Cars" ) 
# most of the costly house owners have 3 cars  
```

```python
sns.relplot(file,x= "Gr Liv Area",y = "SalePrice",col = "Neighborhood",hue = "Overall Cond" ,style =  "Garage Cars" )  
# many of the neighborhoods have good conditioned in low prices
```
